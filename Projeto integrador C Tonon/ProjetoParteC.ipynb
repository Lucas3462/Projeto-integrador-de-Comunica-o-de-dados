## Modelagem com Árvore de Decisão

Nesta etapa do projeto será aplicada uma técnica de Inteligência Artificial para prever a inadimplência dos clientes com base nas variáveis presentes no dataset.

O modelo escolhido foi a **Árvore de Decisão**, que é um algoritmo de aprendizado supervisionado utilizado em problemas de classificação. Esse modelo aprende padrões a partir dos dados históricos e gera regras de decisão que ajudam a identificar se um cliente possui maior ou menor risco de inadimplência.

A variável alvo utilizada será:

**default.payment.next.month**

Essa variável indica se o cliente entrou em inadimplência no mês seguinte, sendo:
- **1** = inadimplente
- **0** = não inadimplente

O objetivo desta etapa é:
- separar os dados em treino e teste
- treinar o modelo de árvore de decisão
- avaliar a acurácia do modelo
- visualizar a árvore gerada
- extrair insights sobre as variáveis mais importantes

```python
# Importação das bibliotecas necessárias
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Separação das variáveis independentes (X) e variável alvo (y)
X = df.drop("default.payment.next.month", axis=1)
y = df["default.payment.next.month"]

# Divisão dos dados em treino e teste (70% treino / 30% teste)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Verificando os tamanhos
print("Treino:", X_train.shape)
print("Teste:", X_test.shape)
```

### Treinamento do Modelo

Após a preparação dos dados, o próximo passo é treinar o modelo de Árvore de Decisão.

O treinamento consiste em fazer com que o algoritmo aprenda padrões a partir dos dados de treino, identificando relações entre as variáveis e a inadimplência dos clientes.

Esse processo permite que o modelo consiga posteriormente fazer previsões sobre novos dados.

```python
X = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Criando o modelo de Árvore de Decisão
modelo = DecisionTreeClassifier(random_state=42)

# Treinando o modelo
modelo.fit(X_train, y_train)

print("Modelo treinado com sucesso!")
```

### Avaliação do Modelo

Após o treinamento do modelo, é necessário avaliar o seu desempenho.

Para isso, utilizamos os dados de teste, que não foram usados durante o treinamento. O modelo irá fazer previsões com base nesses dados e, em seguida, será calculada a **acurácia**.

A acurácia representa a porcentagem de acertos do modelo, ou seja, quantas previsões ele realizou corretamente em relação ao total.

```python
# Fazendo previsões com os dados de teste
y_pred = modelo.predict(X_test)

# Calculando a acurácia
acuracia = accuracy_score(y_test, y_pred)

print("Acurácia do modelo:", acuracia)
```

### Visualização da Árvore de Decisão

Para entender melhor como o modelo toma decisões, é possível visualizar a árvore de decisão gerada.
